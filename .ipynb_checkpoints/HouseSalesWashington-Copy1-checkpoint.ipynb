```python
import pandas as pd
```

```python
#read CSV to DF
houseSalesDF = pd.read_csv('house_sales.csv')
```

```python
#print descriptive and info
print(houseSalesDF.describe())
print(houseSalesDF.info())
```

```python
#Loop through attributes to display boxplots
for attribute in houseSalesDF.columns:
    try:
        boxplot = houseSalesDF.boxplot(column=[attribute])
    except:
        pass
```

```python
#Loop through attributes to display scatter plots
for attribute in houseSalesDF.columns:
    try:
        fig = houseSalesDF.plot.scatter(x='price',y=attribute,c='steelblue')
    except:
        pass
```
